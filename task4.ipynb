```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns_nm = ['Id','Entry','target','text']
```

```python
data = pd.read_csv('twitter_training.csv')
```

```python
data
```

```python
data['Positive'].value_counts(ascending=True)
```

```python
colors =('#808080','#FFA500','#FFA500','#FF00FF')
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Positive', data=data, color='#FF00FF')
plt.title("Count_plot for target data", fontsize=18, c='r')
plt.ylabel("Total Count series", fontsize=14, c='b')
plt.xlabel("Reviews", fontsize=14, c='y')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x='Positive', data=data,  kde= True)
plt.title("Hist_plot for target data", fontsize=18, c='r')
plt.ylabel("Total Count series", fontsize=14, c='b')
plt.xlabel("Reviews", fontsize=14, c='#FFA500')
plt.show()
```

```python
b=data['Borderlands'].value_counts()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=b, data=data)
plt.title("Count_plot for Entry data", fontsize=18, c='r')
plt.ylabel("Total Count series", fontsize=14, c='b')
plt.xlabel("Entry for each Data", fontsize=14, c='#FFA500')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.histplot(x='Borderlands', data=data,  kde= True)
plt.title("Hist_plot for target data", fontsize=18, c='r')
plt.ylabel("Total Count series", fontsize=16, c='b')
plt.xlabel("Target", fontsize=16, c='#FFA500')
plt.show()
```

```python
plt.figure(figsize = (10 , 10))
t = data["Positive"].value_counts()
labels = ["Negative" , "Positive" , "Neutral" , "Irrelevant"]
plt.subplot(2,1,1)
plt.pie(t , labels = labels , autopct = "%1.1f%%")
plt.legend()
plt.title("Pie for Reviews " , fontsize = 10 , c = "k")
```
